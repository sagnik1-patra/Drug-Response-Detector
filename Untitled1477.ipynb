```python
# ============================================
# 🧬 TheraGuide – Visualization & Evaluation Script
# ============================================

import os
import json
import joblib
import yaml
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from tensorflow.keras.models import load_model

# ============================================
# ⚙️ Paths
# ============================================
BASE_DIR = r"C:\Users\sagni\Downloads\Drug Response Detector"
DATA_PATH = os.path.join(BASE_DIR, r"archive\DRP-for-Typhoid-Fever.csv")
MODEL_PKL = os.path.join(BASE_DIR, "theraguide_model.pkl")
MODEL_H5 = os.path.join(BASE_DIR, "theraguide_model.h5")
RECOMM_JSON = os.path.join(BASE_DIR, "recommendations.json")
META_YAML = os.path.join(BASE_DIR, "theraguide_metadata.yaml")

# ============================================
# 📥 Load Dataset
# ============================================
print("[INFO] Loading dataset...")
df = pd.read_csv(DATA_PATH)

# Identify target
target_candidates = [c for c in df.columns if 'response' in c.lower() or
                     'label' in c.lower() or 'outcome' in c.lower() or 'target' in c.lower()]
TARGET = target_candidates[0] if target_candidates else df.columns[-1]
print(f"[INFO] Target detected: {TARGET}")

X = df.drop(columns=[TARGET])
y = df[TARGET]

# Label encode categorical variables
from sklearn.preprocessing import LabelEncoder, StandardScaler
for col in X.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))

if y.dtype == 'object':
    le_y = LabelEncoder()
    y = le_y.fit_transform(y)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split (to match earlier proportions)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)

# ============================================
# 🧩 Load Models
# ============================================
rf = joblib.load(MODEL_PKL)
nn_model = load_model(MODEL_H5)

print("[INFO] Models loaded successfully.")

# ============================================
# 🌲 RandomForest Evaluation
# ============================================
rf_pred = rf.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)
print(f"[RESULT] RandomForest Accuracy: {rf_acc:.4f}")

# ============================================
# 🧠 Neural Network Evaluation
# ============================================
nn_probs = nn_model.predict(X_test)
nn_pred = np.argmax(nn_probs, axis=1)
nn_acc = accuracy_score(y_test, nn_pred)
print(f"[RESULT] Neural Network Accuracy: {nn_acc:.4f}")

# ============================================
# 📊 Plot 1 – Accuracy Comparison
# ============================================
plt.figure(figsize=(6, 4))
bars = plt.bar(['RandomForest', 'NeuralNet'], [rf_acc, nn_acc], color=['#4a90e2', '#50e3c2'])
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() - 0.05,
             f"{bar.get_height():.2f}", ha='center', color='black', fontsize=11)
plt.tight_layout()
plt.show()

# ============================================
# 📊 Plot 2 – Confusion Matrix Heatmap (RandomForest)
# ============================================
cm = confusion_matrix(y_test, rf_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm')
plt.title("Confusion Matrix – Drug Response Classifier (RF)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()

# ============================================
# 📈 Plot 3 – Result Graph (True vs Predicted)
# ============================================
plt.figure(figsize=(10, 4))
plt.plot(y_test[:50], label='True Outcome', marker='o')
plt.plot(rf_pred[:50], label='Predicted Outcome (RF)', marker='x')
plt.title("True vs Predicted Drug Response (First 50 Patients)")
plt.xlabel("Sample Index")
plt.ylabel("Outcome Label")
plt.legend()
plt.tight_layout()
plt.show()

# ============================================
# 📈 Plot 4 – Prediction Confidence Graph
# ============================================
rf_probs = rf.predict_proba(X_test)
confidences = np.max(rf_probs, axis=1)

plt.figure(figsize=(10, 4))
plt.bar(range(len(confidences)), confidences, color='lightgreen')
plt.title("Prediction Confidence per Patient (RandomForest)")
plt.xlabel("Patient Index")
plt.ylabel("Confidence")
plt.tight_layout()
plt.show()

# ============================================
# 📜 Optional: Detailed Classification Report
# ============================================
print("\n[REPORT] RandomForest Classification Report:\n",
      classification_report(y_test, rf_pred))

print("[INFO] Visualization completed successfully.")
```
